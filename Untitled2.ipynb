```python
# ============================================
# BMI Prediction using K-Fold Cross Validation
# ============================================

# Step 1: Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error

# Step 2: Create a Realistic BMI Dataset
# BMI = weight(kg) / (height(m))^2
np.random.seed(42)
heights = np.random.normal(1.65, 0.1, 1000)  # mean height = 1.65m
weights = np.random.normal(65, 15, 1000)     # mean weight = 65kg
bmi = weights / (heights ** 2)

# Combine into a DataFrame
df = pd.DataFrame({'Height(m)': heights, 'Weight(kg)': weights, 'BMI': bmi})

print("Sample of Dataset:")
print(df.head())

# Step 3: Define features and target
X = df[['Height(m)', 'Weight(kg)']]
y = df['BMI']

# Step 4: Initialize model
model = LinearRegression()

# Step 5: 2-Fold Cross Validation
kf2 = KFold(n_splits=2, shuffle=True, random_state=42)
mse_scores_2fold = []

print("\n=== 2-Fold Cross Validation ===")
fold = 1
for train_index, test_index in kf2.split(X):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    mse_scores_2fold.append(mse)
    print(f"Fold {fold} - MSE: {mse:.3f}")
    fold += 1

print(f"Average MSE (2-Fold): {np.mean(mse_scores_2fold):.3f}")

# Step 6: 3-Fold Cross Validation
kf3 = KFold(n_splits=3, shuffle=True, random_state=42)
mse_scores_3fold = []

print("\n=== 3-Fold Cross Validation ===")
fold = 1
for train_index, test_index in kf3.split(X):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    mse_scores_3fold.append(mse)
    print(f"Fold {fold} - MSE: {mse:.3f}")
    fold += 1

print(f"Average MSE (3-Fold): {np.mean(mse_scores_3fold):.3f}")

# Step 7: Visualize predictions
model.fit(X, y)
y_pred_full = model.predict(X)

plt.figure(figsize=(8,6))
plt.scatter(y, y_pred_full, color='teal', alpha=0.6)
plt.plot([y.min(), y.max()], [y.min(), y.max()], color='red', linestyle='--', linewidth=2)
plt.title("BMI Prediction using Linear Regression")
plt.xlabel("Actual BMI")
plt.ylabel("Predicted BMI")
plt.grid(True)
plt.show()
```
